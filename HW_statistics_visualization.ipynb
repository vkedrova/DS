```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gas_data = pd.read_csv(r'C:\Users\vkedrova\Desktop\personal\Python\Data_Sciense_course\Statistics\Gas sales with temperature.csv')
gas_data.head()
```

```python
gas_data.describe()
```

Похоже, что в данных нет пропусков, что не исключает нулевых значений

```python
gas_data.isna().sum()
```

Если брать суммарный объем потребления газа во всех городах, получаем правосторонее распределение, можно предположить, городов с меньшим потреблением в выборке больше

```python
sns.histplot(data=gas_data, x= "Sum", bins=40, kde=True)
plt.title("Total Gas consumption")
plt.ylabel("Amount")
plt.show()
```

В суммарном потреблении не находим выбросов, возможно, суммарные данные слишком разнородны для такого анализа

```python
plt.boxplot(data=gas_data, x='Sum')
plt.title("Total Gas consumption outliers")
plt.show()
```

Данные по суммарному потреблению показывают логичные тенденции - с годами объем потребления газа увеличивается, температура выше в летние месяцы и с понижением температуры повышается потребление газа

```python
gas_data.groupby('Year')['Sum'].sum().plot(kind = 'bar')
plt.show()
```

```python
gas_data.groupby('Month')['Temperature'].mean().plot(kind = 'bar')
plt.show()
```

```python
sns.scatterplot(x = gas_data['Temperature'], y = gas_data['Sum'])
plt.show()
```

Посмотрим на города в отдельности. У некоторых более выражена правостороняя тенденция, как будто в этих городах потребляют меньше газа в целом по сравнении с другими - возможно, это связано с плотностью населения или температурой (но у нас нет этих данных в разрезах городов)

```python
gas_data.hist(bins=25, figsize=(15, 25), layout=(-1, 5), edgecolor="black")
plt.tight_layout()
```

Большинство городов, имеющих выраженное правосторонее распределение, также показывают заметный объем максимальных экстремумов 

```python
gas_data.plot(kind = 'box', subplots=True, figsize=(10, 15), layout=(-1, 5))
plt.tight_layout()
```

Для некоторых городов смещение вызвано тем, что наблюдения начинаются позднее 2000 года, для сравнительного анализа по всем периодам их лучше не использовать - могут искажать среднее и/или другие показатели

```python
gas_data.groupby('Year')['Jeju', 'Jeollanamdo', 'Sejong', 'Gangwondo'].mean().plot(kind = 'bar', subplots=True)
plt.tight_layout()
```

В 2020 году присутствуют данные по всем странам, построим рейтинг потребления в 2020 году, более густонаселенные регионы, ожидаемо, потребляют больше газа

```python
gas_data_countries = gas_data.melt(id_vars=['Year', 'Month', 'Temperature', 'Sum'], var_name = 'Country', value_name = 'Amount')
```

```python
gas_data_countries.loc[gas_data_countries['Year'] == 2020].groupby('Country')['Amount'].sum().sort_values().plot(kind = 'barh')
plt.show()
```
