```python
def traffic_type(df):
    if ('yandex' in df['traffic_source']) or ('google' in df['traffic_source']):
        df['traffic_type'] = 'organic'
    elif ('Russia' in df['region']) and (('paid' in df['traffic_source']) or ('email' in df['traffic_source'])):
        df['traffic_type'] = 'ad'
    elif ('Russia' not in df['region']) and (('paid' in df['traffic_source']) or ('email' in df['traffic_source'])):
        df['traffic_type'] = 'other'
    else:
        df['traffic_type'] = df['traffic_source']
        
    return df         
```

```python
log = log.apply(traffic_type, axis = 1)
log.groupby('traffic_type').count()
```

# Задание 3

```python
ratings = pd.read_csv(r'C:\Users\vkedrova\Desktop\personal\Python\Data_Sciense_course\доп_материалы_к_пандас_1_и_2\ml-latest-small\ratings.csv')
ratings.head()
```

```python
counts = ratings.groupby('userId').count().reset_index().drop(['movieId', 'timestamp'], axis = 1).rename(columns = {'rating' : 'counts'})
counts.head()
```

```python
def lifetime (df):
    min_tsmp = df['timestamp'].min()
    max_tsmp = df['timestamp'].max()
    df['lifetime'] = max_tsmp - min_tsmp
    return df
```

```python
ratings = ratings.groupby('userId').apply (lifetime)
ratings.groupby('userId').mean()
```

```python
lifetime = counts.merge(ratings, on = 'userId', how = 'left')
lifetime.loc[lifetime.counts > 100]
```
