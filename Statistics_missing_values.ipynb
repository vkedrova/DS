# Задание 2. Первичное изучение данных

```python
#Базовые статистики
key_hourse_data.describe()
```

```python
#Чаще всего лошадки выживают)
key_hourse_data['outcome'].mode()
```

```python
#Для данных, которым неактуален расчет среднего или выбросов может быть полезен pivot, например, распределение исхода в зависимости от наличия операции
pd.pivot_table(key_hourse_data, values = 'Age',  index = 'surgery?', columns = 'outcome', aggfunc = 'sum')
```

```python
#По условиям описания нормальные показатели 8-10, т.к. 8 это минимальное значение ряда, можно найти отклонения выше 10
resp_higher_outliers = key_hourse_data[key_hourse_data['respiratory rate'] > 10]
resp_lower_outliers.groupby('respiratory rate').count()
```

```python
#Поиск выбросов для rectal temperature
q1 = key_hourse_data['rectal temperature'].quantile(0.25)
q3 = key_hourse_data['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

remove_outliers = key_hourse_data[key_hourse_data['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)]
```

```python
#Судя по полученным минимальным отличиям в ряду практически нет выбросов, данные имеют высокую плотность
print(key_hourse_data['rectal temperature'].mean())
print(remove_outliers['rectal temperature'].mean())
```

```python
#Поиск выбросов для pulse
q1 = key_hourse_data['pulse'].quantile(0.25)
q3 = key_hourse_data['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound_pulse = q1 - (1.5 * iqr)
upper_bound_pulse = q3 + (1.5 * iqr)

remove_outliers_pulse = key_hourse_data[key_hourse_data['pulse'].between(lower_bound_pulse, upper_bound_pulse, inclusive=True)]
```

```python
#В показателях пульса средние значения отличаются более заметно, медианные равны, выбросы имеют скорее единичный характер
print(key_hourse_data['pulse'].mean())
print(remove_outliers_pulse['pulse'].mean())

print(key_hourse_data['pulse'].median())
print(remove_outliers_pulse['pulse'].median())
```

```python
#Проверка количества выбросов pulse
key_hourse_data[~key_hourse_data['pulse'].between(lower_bound_pulse, upper_bound_pulse, inclusive=True)]
```

# Задание 3. Работа с пропусками

```python
#Исходя из кол-ва пропусков заполнять их имеет смысл в rectal temperature, pulse, respiratory rate, pain и temperature of extremities
(key_hourse_data.isna().mean() * 100).round(1)
```

```python
#Данные о том была ли операция и чем закончилось наблюдение спрогнозировать сложно исходя из имеющихся данных
key_hourse_data['surgery?']  = key_hourse_data['surgery?'].fillna('no_data')
key_hourse_data['outcome']  = key_hourse_data['outcome'].fillna('no_data')
```

```python
#В описании данных есть пометка, что respiratory rate имеет сомнительную аналитическую ценность ввиду большого количества флуктуаций
key_hourse_data['respiratory rate']  = key_hourse_data['respiratory rate'].fillna('no_data')
```

```python
# Т.к. для rectal temperature и pulse данные имеют минимальное кол-во выбросов, можно заполнить их средним и медианой соответственно с минимальным риском искажения
key_hourse_data['rectal temperature'] = key_hourse_data['rectal temperature'].fillna(key_hourse_data['rectal temperature'].mean())
key_hourse_data['pulse'] = key_hourse_data['pulse'].fillna(key_hourse_data['pulse'].median())
```

```python
#Данные в столбцах pain и temperature of extremities носят номинальный характер, при необходимости можно попробовать заполнить модой
```

```python
#Наиболее частые значения температуры конечностей в заависимости от исхода
key_hourse_data.groupby(['outcome'])['temperature of extremities'].agg(pd.Series.mode)
```

```python
#Наиболее частые значения температуры конечностей в заависимости от факта операции
key_hourse_data.groupby(['surgery?'])['pain'].agg(pd.Series.mode)
```
