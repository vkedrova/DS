```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/Users/valeria/Downloads/Almagel_market_BRANDS.csv', sep = ';')
df.set_index('Year', inplace = True)
df['Total_year'] = df.sum(axis = 1)
df.head()
```

```python
df['Total_year'].plot(kind = 'area', title = 'Total market dynamic', xlabel = 'Year', ylabel = 'mln.RUR')
#Market growing up to 2020 but stagnating in 2021
```

```python
tdf = df.T.sort_values(by = 2021, ascending = False)
tdf.drop(index = 'Total_year', inplace = True)
tdf.head().style.bar()
#Top 5 Market contributors are Almagel, Phosphalugel, Gastal, Maalox, Rennie
```

```python
top5 = df[['Almagel a', 'Phosphalugel', 'Gastal', 'Maalox', 'Rennie mint']]
top5.plot(title = 'Top 5 market dynamic', ylabel = 'mln.RUB')
plt.legend(loc='upper left')
#Phosphalugel significant decrease in 2021 became main Market decliner
```
