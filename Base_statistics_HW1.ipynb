```python
import pandas as pd
import numpy as np
```

```python
df_ = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv', names = [i for i in range(0,28)], na_values = '?')
df_.head()
```

```python
df = df_[[0, 1, 3, 4, 5, 6, 10, 22]]
df.columns = ['surgery', 'age', 'rec_temp', 'pulse', 'resp_rate', 'temp_of_ext', 'pain', 'outcome']
df.describe()
```

```python
df['surgery'].mode()
df['surgery'].value_counts()
#Оперированные животные встречаются чаще
df['surgery'] = df['surgery'].fillna(df['surgery'].mean())
```

```python
print(df['outcome'].mode())
print(df['outcome'].value_counts())
#Выжившие лошадки встречаются заметно чаще, можно попробовать заполнить единственный пробел модой
df['outcome'] = df['outcome'].fillna(df['outcome'].mode())
```

```python
print(df['rec_temp'].mean())
print(df['rec_temp'].median())
#Медиана и среднее практически совпадают, выбросы скорее всего незначиельные
```

```python
q1_rec_temp = df['rec_temp'].quantile(0.25)
q3_rec_temp = df['rec_temp'].quantile(0.75)
qrt = q3_rec_temp - q1_rec_temp
rec_lower_out = q1_rec_temp - (qrt *1.5)
print(rec_lower_out)
rec_higher_out = q3_rec_temp + (qrt *1.5)
print(rec_higher_out)
#Выбросы от среднего кажутся не драматичными, пропуски можно заполнить и средним
df['rec_temp'] = df['rec_temp'].fillna(df['rec_temp'].mean())
```

```python
print(df['pulse'].mean())
print(df['pulse'].median())
print(df['pulse'].unique())
#Медиана заметно отличается от среднего - возможны заметные выбросы. Данные достаточно разнородны
```

```python
q1_pulse = df['pulse'].quantile(0.25)
q3_pulse = df['pulse'].quantile(0.75)
qrt_pulse = q3_pulse - q1_pulse
pulse_lower_out = q1_pulse - (qrt_pulse * 1.5)
print(pulse_lower_out)
pulse_higher_out = q3_pulse + (qrt_pulse * 1.5)
print(pulse_higher_out)
#Межквартильный размах получается большим, похоже, для этого ряда данных считать выбросы так не подходит. Раз пульс настолько индивидуальная мера - лучше бы вообще ничем не заполнять пропуски, но если очень нужно - наверное, лучше медианой
df['pulse'] = df['pulse'].fillna(df['pulse'].median())
```

```python
q1_respr = df['resp_rate'].quantile(0.25)
q3_respr = df['resp_rate'].quantile(0.75)
qrt_respr = q3_respr - q1_respr
respr_lower_out = q1_respr - (qrt_respr * 1.5)
print(respr_lower_out)
respr_higher_out = q3_respr + (qrt_respr * 1.5)
print(respr_higher_out)
df['resp_rate'].unique()
#Данные также достаточно разнородны, ситуация похожа на поле pulse
df['resp_rate'] = df['resp_rate'].fillna(df['resp_rate'].median())
```

```python
print(df['temp_of_ext'].value_counts())
print(df['temp_of_ext'].mode())
#Хотя можно выделить моду, данные кажутся разрозненными
print(df.groupby(['temp_of_ext']).value_counts())
#Довольно сложно выделить закономерность для заполнения пропусков на базе анализа взаимосвязи с другими данными. Если это необходимо - можно попробовать заполнить модой
df['temp_of_ext'] = df['temp_of_ext'].fillna(df['temp_of_ext'].mode())
```

```python
print(df['pain'].value_counts())
print(df['pain'].mode())
#Хотя можно выделить моду, данные кажутся разрозненными
print(df.groupby(['pain']).value_counts())
#Показатель достаточно субъективный и мало влияет на диагностику. Я бы не заполняла это поле
```
